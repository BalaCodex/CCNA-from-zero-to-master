     Box plot


```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------------- DATA ----------------
df = pd.DataFrame({
    "Player": ["Virat Kohli", "Rohit Sharma", "Joe Root", "Babar Azam", "Kane Williamson",
               "Steve Smith", "David Warner", "Shubman Gill", "KL Rahul", "AB de Villiers"],
    "HomeTeam": ["India", "India", "England", "Pakistan", "New Zealand",
                 "Australia", "Australia", "India", "India", "South Africa"],
    "Age": [35, 36, 33, 29, 33, 34, 37, 25, 31, 39],
    "Runs": [12898, 10709, 6207, 5409, 6554, 4932, 5455, 2999, 1986, 9577]
})

# --------------------------------------------------
# 1 SCATTER PLOT  Numerical vs Numerical
# Question: "How are Age and Runs related?"
plt.figure()
plt.scatter(df["Runs"], df["Age"])
plt.xlabel("Runs")
plt.ylabel("Age")
plt.title("Scatter: Age vs Runs")
plt.show()

# --------------------------------------------------
# 2 BAR PLOT  Categorical vs Numerical
# Question: "Average runs per HomeTeam?"
home_avg = df.groupby("HomeTeam")["Runs"].mean()

plt.figure()
plt.bar(home_avg.index, home_avg.values)
plt.xticks(rotation=45)
plt.ylabel("Average Runs")
plt.title("Bar: Average Runs by HomeTeam")
plt.show()

# --------------------------------------------------
# 3 HISTOGRAM  Single Numerical (distribution)
# Question: "How are Runs distributed?"
plt.figure()
plt.hist(df["Runs"], bins=5)
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.title("Histogram: Runs Distribution")
plt.show()

# --------------------------------------------------
# 4 BAR (COUNT)  Categorical vs Categorical
# Question: "How many players per team?"
team_count = df["HomeTeam"].value_counts()

plt.figure()
plt.bar(team_count.index, team_count.values)
plt.xticks(rotation=45)
plt.ylabel("Number of Players")
plt.title("Bar: Players per HomeTeam")
plt.show()

# --------------------------------------------------
# 5 BOX PLOT  Numerical distribution by Category
# Question: "How do Runs vary by team?"
teams = df["HomeTeam"].unique()
runs_by_team = [df[df["HomeTeam"] == t]["Runs"] for t in teams]

plt.figure()
plt.boxplot(runs_by_team, labels=teams)
plt.xticks(rotation=45)
plt.ylabel("Runs")
plt.title("Box Plot: Runs by HomeTeam")
plt.show()
```
